```python
import os
```

```python
folders = [f for f in os.listdir() if os.path.isdir(f) and "results" in f]
```

```python
sample_files = [f"{f}/posterior_sample.txt" for f in folders]
```

```python
fig, axes = plt.subplots(2,2, figsize=(20,10))

axes = np.hstack(axes)

for i, f in enumerate(sample_files):
    sample = np.loadtxt(f, skiprows=2)
    u, c = np.unique(sample[:,9], return_counts=True)
    axes[i].bar(u,c)
    axes[i].set_title(f"Components for {f}")
```

> This looks like a Poisson distribution. At least for the max 100 components version. The others look more Gaussian like. Overall looks like they have similar means if we do not count the max 10 components version. Seems like it really prefers to have more components just to make sure it samples everything.
